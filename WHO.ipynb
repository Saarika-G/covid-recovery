# Inheritance

```python
def WHO(a):
    '''Inheritance'''
    return a**a
print (WHO.__doc__)
  
# parent class
class WHO(object):
  
    # __init__ is known as the constructor
    def __init__(self,populationinthousands):
        self.populationinthousands = populationinthousands
  
    def display(self):
        print(self.populationinthousands)
          
    def details(self):
        print("Total population is: {}".format(populationinthousands))
      
# child class
class COVID(WHO):
    def __init__(self,Country,populationinthousands,Confirmed,Deaths,Recovered,Active,Newcases,Newdeaths,Newrecovered):
        self.Country = Country
        self.populationinthousands = populationinthousands
        self.Confirmed = Confirmed
        self.Deaths = Deaths
        self.Recovered = Recovered
        self.Active = Active
        self.Newcases = Newcases
        self.Newdeaths = Newdeaths
        self.Newrecovered = Newrecovered
        
        # invoking the __init__ of the parent class
        WHO.__init__(self,populationinthousands)
          
    def details(self):
        print("Country: {}".format(self.Country))
        print("Total population in thousands is: {}".format(self.populationinthousands))
        print("ConfirmedCases: {}".format(self.Confirmed))
        print("DeathsReported: {}".format(self.Deaths))
        print("Recovered: {}".format(self.Recovered))
        print("ActiveCases: {}".format(self.Active))
        print("NewCases: {}".format(self.Newcases))
        print("Newdeaths: {}".format(self.Newdeaths))
        print("Newrecovered: {}".format(self.Newrecovered))
  
  
# creation of an object variable or an instance
a = COVID('Afghanistan', 26088, 36263, 1269, 25198, 9796, 106, 10, 18)
  
# calling a function of the class Person using
# its instance
a.display()
a.details()
```

```python
#To calculate RMSE Values
def RMSE(pred,actual):
    return np.sqrt(np.mean((actual-pred)**2))
```

```python
#predict y values for train data 
train_yhat = model1.predict(train_data)
train_y = train_data.Recovered_Percent
RMSE_train = RMSE(train_yhat, train_y)
print("RMSE value of train data is",round(RMSE_train,4))
```

```python
#predict y values for test data
test_yhat = model1.predict(test_data)
test_y = test_data.Recovered_Percent
RMSE_test = RMSE(test_yhat,test_y)
print("RMSE value of test data is",round(RMSE_test,4))
```
